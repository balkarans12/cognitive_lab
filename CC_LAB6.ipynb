```python
#QUESTION 1
import numpy as np
import matplotlib.pyplot as p
m=int(input("Enter the value for m:"))
t=np.linspace(-10,10,50)
u1=m*t**2
u2=m*np.sin(t)
p.plot(t,u1,color='r',linestyle='dashed', label=f'{m}x^2')
p.plot(t,u2,color='g',linestyle='solid', label=f'{m}sin(x)')
p.legend()
p.grid(True)
p.title(f'Plots of y={m}x^2 and y={m}sin(x)')
p.show()
```

```python
#QUESTION 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
m=int(input("Enter the value for m:"))
t=np.linspace(-10,10,50)
u1=m*t**2
u2=m*np.sin(t)
p.plot(t,u1,color='r',linestyle='dashed', label=f'{m}x^2')
p.plot(t,u2,color='g',linestyle='solid', label=f'{m}sin(x)')
p.legend()
p.grid(True)
p.title(f'Plots of y={m}x^2 and y={m}sin(x)')
p.show()
s=['Math','Science','English','History','Geography']
v=[78,85,90,88,76]
d=pd.DataFrame({'Subject':s,'Score':v})
a=d.plot(x='Subject',y='Score',kind='bar',color=['r','g','b','y','m'])
for i in a.containers:
    a.bar_label(i)
p.title('Subject Scores')
p.xlabel('Subjects')
p.ylabel('Scores')
p.grid()
p.show()
```

```python
#QUESTION 3
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
np.random.seed(102317275)
data=np.random.randn(50)
fig,ax=p.subplots(2,2)
ax[0,0].plot(np.cumsum(data),color='b',marker='o')
ax[0,0].set_title('Line Plot Showing Cumulative Sum')
ax[0,0].grid(alpha=0.4)

ax[0,1].scatter(range(50),data+np.random.randn(50)*0.2,color='r')
ax[0,1].set_title('Scatter Plot')
ax[0,1].grid(alpha=0.4)
p.show()
```
