```python
#question 1
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = np.sum(gfg)

row_wise_sum = np.sum(gfg, axis=1)

column_wise_sum = np.sum(gfg, axis=0)

print("Sum of all elements:", total_sum)
print("Row-wise sum:\n", row_wise_sum)
print("Column-wise sum:\n", column_wise_sum)
```

```python
#question 2(a)
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)

sorted_indices = np.argsort(array)

smallest_elements = np.sort(array)[:4]

largest_elements = np.sort(array)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_elements)
print("5 largest elements:", largest_elements)
```

```python
#question 2(b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == array.astype(int)]

float_elements = array[array != array.astype(int)]

print("Integer elements only:", integer_elements)
print("Float elements only:", float_elements)
```

```python
#question 3(a)

initials = "A B"  # Replace with your actual initials
ascii_sum = sum(ord(char) for char in initials if char.isalpha())  # Compute ASCII sum (X)
X = ascii_sum

sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales Data:", sales)

# 3(b)
tax_rate = ((X % 5) + 5) / 100
tax_amount = sales * tax_rate
print("Tax Amount on Sales:", tax_amount)

# 3(c)
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)  # 5% for < X+100, 10% for >= X+100
print("Discounted Sales Data:", discounted_sales)

# 3(d)
weeks = 3
weekly_sales = np.vstack([discounted_sales] * weeks)  # Stack 3 times to create a 3Ã—5 matrix
print("Weekly Sales (Before Increase):\n", weekly_sales)

week_factors = (1.02) ** np.arange(weeks).reshape(-1, 1)  # 1.02^week index
adjusted_weekly_sales = weekly_sales * week_factors  # Broadcasting multiplication
print("Weekly Sales (After 2% Increase per Week):\n", adjusted_weekly_sales)
```

```python
# question (4)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2               # Y = x^2
y2 = np.sin(x)          # Y = sin(x)
y3 = np.exp(x)          # Y = e^x
y4 = np.log(np.abs(x) + 1)  # Y = log(|x| + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label=r'$y = x^2$', linestyle='--', color='b')
plt.plot(x, y2, label=r'$y = \sin(x)$', linestyle='-', color='g')
plt.plot(x, y3, label=r'$y = e^x$', linestyle=':', color='r')
plt.plot(x, y4, label=r'$y = \log(|x| + 1)$', linestyle='-.', color='m')

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plots of Various Functions")
plt.legend()
plt.grid(True)

plt.show()
```
